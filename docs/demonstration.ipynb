# Gapminder package demonstration

**Jupyter Notebooks** are quite useful to do exploratory data analyses. It's capacity to render Markdown and execute code in the same document make it perfect to combine explanatory text with Python code. In this case, we are going to use this notebook to demonstrate some of the functions in the `gapminder.py` module.

Because we are in the same directory as the module, we can import it directly:

```python
import gapminder
```

The `load_gapminder_all()` function loads the data (located in the `gapminder_all.csv` file):

```python
df = gapminder.load_gapminder_all()
```

The `create_df_views()` function splits the data so we can plot it. It returns four dataframes: one for each variable (GDP per capita, life expectancy, and population), and a melted dataframe with all the data.

```python
gdp, life, pop, cat = gapminder.create_df_views(df)
```

The `plot_world_development_by_year()` creates a nice scatter plot. It receives the `cat` dataframe as input.

```python
gapminder.plot_world_development_by_year(cat)
```

The default year is `'2007'`, but we can choose a different one:

```python
gapminder.plot_world_development_by_year(cat, year='1957')
```

The function returns a helpful error message if there is no data available for that year:

```python
gapminder.plot_world_development_by_year(cat, year='2021')
```
